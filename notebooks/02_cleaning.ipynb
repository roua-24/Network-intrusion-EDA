# 02 — Data Cleaning & Preprocessing
Encoding categorical variables, feature selection, normalization.

```python
import pandas as pd
import numpy as np

columns = [
    'duration', 'protocol_type', 'service', 'flag', 'src_bytes',
    'dst_bytes', 'land', 'wrong_fragment', 'urgent', 'hot',
    'num_failed_logins', 'logged_in', 'num_compromised', 'root_shell',
    'su_attempted', 'num_root', 'num_file_creations', 'num_shells',
    'num_access_files', 'num_outbound_cmds', 'is_host_login',
    'is_guest_login', 'count', 'srv_count', 'serror_rate',
    'srv_serror_rate', 'rerror_rate', 'srv_rerror_rate', 'same_srv_rate',
    'diff_srv_rate', 'srv_diff_host_rate', 'dst_host_count',
    'dst_host_srv_count', 'dst_host_same_srv_rate',
    'dst_host_diff_srv_rate', 'dst_host_same_src_port_rate',
    'dst_host_srv_diff_host_rate', 'dst_host_serror_rate',
    'dst_host_srv_serror_rate', 'dst_host_rerror_rate',
    'dst_host_srv_rerror_rate', 'label'
]

DATA_PATH = r"C:\Users\THINKBOOK\Network-intrusion-EDA\data\raw\kddcup.data_10_percent"

df = pd.read_csv(DATA_PATH, names=columns)
print("Dataset loaded successfully.")
print(f"Shape: {df.shape}")
```

```python
# Null values check
print(f"Null values: {df.isnull().sum().sum()}")
```

```python
# Null values check
print(f"Null values: {df.isnull().sum().sum()}")
```

```python
# Binary label : normal=0, anomaly=1
df['is_anomaly'] = df['label'].apply(lambda x: 0 if x == 'normal.' else 1)
print("Label distribution:")
print(df['is_anomaly'].value_counts())
```

```python
# Drop useless column (always 0 in this dataset)
df.drop(columns=['num_outbound_cmds'], inplace=True)
print("Dropped: num_outbound_cmds")
print(f"New shape: {df.shape}")
```

```python
# Encode categorical columns
categorical_cols = ['protocol_type', 'service', 'flag']
df_encoded = pd.get_dummies(df, columns=categorical_cols)
print(f"Shape after encoding: {df_encoded.shape}")
```

```python
# Numerical features only
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
df_num = df[numeric_cols]
print(f"Numerical features: {len(numeric_cols)}")
df_num.describe().T
```

```python
import os

# Create directory if it doesn't exist
os.makedirs(r"C:\Users\THINKBOOK\Network-intrusion-EDA\data\processed", exist_ok=True)

# Save cleaned dataset
SAVE_PATH = r"C:\Users\THINKBOOK\Network-intrusion-EDA\data\processed\kdd_clean.csv"
df.to_csv(SAVE_PATH, index=False)
print(f"Cleaned dataset saved to: {SAVE_PATH}")
```

- num_outbound_cmds dropped : constant 0 across all rows
